```python

# 执行查询并显示完整结果
df = run_query("""
    MATCH (n)
    RETURN labels(n)[0] AS node_type, count(*) AS count
    ORDER BY count DESC
""")
def run_query(query, params=None):
    with driver.session() as session:
        result = session.run(query, params)
        return pd.DataFrame([dict(record) for record in result])

# 查询所有节点类型（无LIMIT限制）
query = """
MATCH (n)
RETURN labels(n)[0] AS node_type, count(*) AS count
ORDER BY count DESC
"""  # 已移除 LIMIT 10

df = run_query(query)
df.head(13)
```
